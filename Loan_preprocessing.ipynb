```python
import pandas as pd
```

```python
train_data = pd.read_csv('train_data.csv')
```

```python
train_data
```

```python
df = train_data.copy()
```

```python
pd.options.display.max_columns = None
pd.options.display.max_rows = None
```

```python
display(df)
```

```python
df.info()
```

```python
df1 = df.drop(['Loan_ID'], axis = 1)
```

```python
df1
```

```python
df1.info()
```

```python
df2 = df1.dropna()
```

```python
df2
```

```python
df2.info()
```

```python
df3 = df2.copy()
```

```python
df3['Gender'] = df3['Gender'].map({'Male':1, 'Female':2})
df3['Married'] = df3['Married'].map({'No':0, 'Yes':1})
df3['Dependents'] = df3['Dependents'].map({'0':0, '1':1, '2':2, '3+':3})
df3['Education'] = df3['Education'].map({'Not Graduate':0, 'Graduate':1})
df3['Self_Employed'] = df3['Self_Employed'].map({'No':0, 'Yes':1})
df3['Property_Area'] = df3['Property_Area'].map({'Rural':1, 'Semiurban':2, 'Urban':3})
df3['Loan_Status'] = df3['Loan_Status'].map({'N':0, 'Y':1})
```

```python
df3
```

```python
df3['CoapplicantIncome'] = df3['CoapplicantIncome'].round(1)
```

```python
df3['ApplicantIncome'] = df3['ApplicantIncome'].astype(float)
```

```python
df3['Credit_History'] = df3['Credit_History'].astype(int)
```

```python
df3
```

```python
df3.head()
```

```python
df_preprocessed = df3
df_preprocessed
```

```python
df_preprocessed.info()
```

```python
df_preprocessed.to_csv('Loan_Status_Preprocessed.csv', index=False)
```
